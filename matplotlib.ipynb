```python
x=[1,2,3,4,5,6,7]
y1= []
for i in x:
    y1.append(i**2)
print(y1)
plt.plot(x,y1)
plt.scatter(x,y1)  #plt.show() is optional.
```

```python
y2=[]
for i in x:
    y2.append(i**3)
print(y2)
plt.plot(x,y2)
plt.scatter(x,y2)
```

```python
y3=[]
for i in x:
    y3.append(i**(1/2))
print (y3)
plt.plot(x,y3)
plt.scatter(x,y3)
plt.xlabel('integers')
plt.ylabel('square roots')
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y1,c='orange')
plt.scatter(x,y2,c='cyan')
plt.scatter(x,y3,c='green')
plt.title('all together')
plt.plot(x,y1,c='green',label='squares')
plt.plot(x,y2,c='blue',label='cubes')
plt.plot(x,y3,c='red',label='sqaure roots')
plt.legend()
plt.show()
```
