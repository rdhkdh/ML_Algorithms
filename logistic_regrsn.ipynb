To predict whether tourist will visit London based on temperature there(t) and ticket price(p)  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
t=np.random.randint(0,40,30)
p=np.random.randint(100,900,30)
print(t)
print(p)
```

```python
df1=pd.DataFrame({ 'temperature': t, 'price': p })
df1
```

```python
plt.scatter(t,p)
plt.xlabel('temperature')
plt.ylabel('price')
```

```python
df1['visiting']=df1['price'] #just creating a column
df1['visiting'] = df1.apply(lambda x: 1 if (6*x['temperature']+2*x['price']<1000) else 0 , axis=1)
df1
```
